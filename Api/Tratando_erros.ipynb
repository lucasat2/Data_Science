# Tratando erros e situações comuns

```python

import requests

url = "https://jsonplaceholder.typicode.com/posts/1"
resposta = requests.get(url)

if resposta.status_code == 200:
    dados = resposta.json()
    print("Título:", dados["title"])
else:
    print("Falha ao buscar dados. Código:", resposta.status_code)
```

Você pode generalizar:

- `2xx` → sucesso.
- `4xx` → erro do cliente (URL errada, parâmetro inválido, não autorizado).
- `5xx` → erro no servidor (problema do lado da API).
