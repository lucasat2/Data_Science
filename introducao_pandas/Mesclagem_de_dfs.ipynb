```python
import pandas as pd
import numpy as np


data = {
	'Escola': ['A', 'B', 'A', 'B', 'A' , 'B' , 'A', 'B'],
	'Student ID': [1001, 1002, 1003, 1004, 1005, 1006,1007, 1008],
	'Matematica': np.random.randint(60,100,8),
	'Ingles': np.random.randint(60,100,8),
	'Historia': np.random.randint(60,100,8)
	}
```

```python
df = pd.DataFrame(data)
```

```python
# Alteramos os 2 ultimos IDS de proposito

sports_data = {
  'Student ID': [1001, 1002, 1003, 1004, 1005,
  1006, 1009, 1010],
  'Esportes': ['Futebol', 'Basquete', 'Volei', 'Futebol','Volei', 'Futebol', 'Basquete','Volei'
]}
```

```python
sports_df = pd.DataFrame(sports_data)
```

```python
sports_df
```

```python
pd.merge(df, sports_df, on='Student ID')
```

```python
# Perceba que somente os ids comuns entre os 2 dfs foram mesclados
```

```python
# Para unir os diferentes precisamos usar o que se segue abaixo
```

```python
merged_outer = pd.merge(df, sports_df, on='Student ID', how='outer')
```

```python
merged_outer
```

```python
# Concatenar dataframes
```

```python
data2 = {
	'Escola': ['A', 'B', 'A', 'B', 'A' , 'B' , 'A', 'B'],
	'Student ID': [1009,1010,1011,1012,1013,1014,1015,1016],
	'Matematica': np.random.randint(60,100,8),
	'Ingles': np.random.randint(60,100,8),
	'Historia': np.random.randint(60,100,8)
	}

df2 = pd.DataFrame(data2)
```

```python
concatenado = pd.concat([df,df2], ignore_index=True)
```

```python
concatenado
```
