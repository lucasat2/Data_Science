```python
import numpy as np
import pandas as pd
```

```python
# Crie um intervalo de datas para um ano
dates_rng= pd.date_range(start = '1/1/2018', end='12/31/2018', freq ='D')
```

```python
# Gerando flutuações aleatórias em torno do preço inicial da ação, base em 100 reais
stock_data = 100 + np.cumsum(np.random.randn(365)* 2)
```

```python
# Criando um Dataframe
df = pd.DataFrame(stock_data, index=dates_rng, columns=['Preco'])
```

```python
df
```

```python
# Preço médio da ação por mes

mensal = df.resample('ME').mean()
```

```python
mensal
```

```python
# preço no dia seguinte

#criando uma nova coluna com o preço deslocado de 1 dia

df['Preco_dia_seguinte'] = df['Preco'].shift(1)

#variacao diaria do preço
df['Mudanca diaria'] = df['Preco'] - df['Preco_dia_seguinte']
```

```python
df
```

```python
# Janelas Rolantes - Ajudam a observar a tendencia em um tempo

# Calcular a media movel em 7 dias.

df_rolling = df.rolling(7).mean()


# os 7 primeiros valores darao nulo pois ainda nao temos dados anteriores
```
