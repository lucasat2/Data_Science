## **7. Enviando parâmetros na URL com `params`**

Muitas APIs usam **parâmetros de consulta** (query string) para filtrar ou personalizar os dados retornados. Exemplo de URL com parâmetros:

https://api.exemplo.com/produtos?categoria=livros&ordenar=preco

Você não precisa montar essa URL manualmente. O requests faz isso pra você usando o argumento params.

```python
import requests  # biblioteca para fazer requisições HTTP
```

```python
url = "https://jsonplaceholder.typicode.com/comments"

parametros = {
    "postId": 1  # queremos comentários apenas do post com id=1
}

resposta = requests.get(url, params=parametros)

print("URL final:", resposta.url)  # veja como ficou a query string
print("Status code:", resposta.status_code)

comentarios = resposta.json()

print("Quantidade de comentários retornados:", len(comentarios))
print("Primeiro comentário:", comentarios[0])
print("Segundo comentário:", comentarios[1])
```
