```python
import numpy as np
import pandas as pd

data = {
'A': [1,6,np.nan,8,5],
'B':[np.nan,6,3,4,5],
'C': [1,8,3,np.nan,np.nan]
}
df = pd.DataFrame(data)
df

# df com notas de alunos
# 3 disciplinas (A,B,C), a média é 6
# nulo nao fez, 4 reprovado e 5 recuperaçao
```

```python
def calculaNotas(nota):

	if pd.isna(nota):
		return 'Nao fez a prova'
	elif nota <= 4:
		return 'Reprovado'
	elif nota < 6 and nota >= 5:
		return 'Recuperação'
	else:
		return 'Aprovado'

df_com_notas = df.map(calculaNotas)
df_com_notas
```
