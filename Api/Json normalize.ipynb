```python
usuarios = [
    {
        "id": 1,
        "nome": "Ana",
        "endereco": {
            "cidade": "Juiz de Fora",
            "estado": "MG"
        },
        "planos": ["básico", "data-science"]
    },
    {
        "id": 2,
        "nome": "Bruno",
        "endereco": {
            "cidade": "São Paulo",
            "estado": "SP"
        },
        "planos": ["premium"]
    },
    {
        "id": 3,
        "nome": "Carla",
        "endereco": {
            "cidade": "Rio de Janeiro",
            "estado": "RJ"
        },
        "planos": ["básico"]
    }
]
```

# Use json_normalize para transformar usuarios em um DataFrame.



```python
from pandas import json_normalize

df = json_normalize(usuarios)
```

# Verifique quais colunas foram criadas.

```python
print(df.columns)
```

# Crie um DataFrame df_simplificado com as colunas:

```python
df_simplificado = df.rename(columns= {

    'endereco.cidade' : 'cidade',
    'endereco.estado' : 'estado',
})
```

```python
print(df_simplificado.columns)
```
