# Agrupamento 

```python
# Cenario 1 : Agrupando por origem e destino 

grouped = df.groupby(['Origin', 'Destination'])

# Aplicar e agregar uma funcao ao grupo 

aggregated = grouped.agg({'Passengers':'sum'})

#Selecione os valores dessa tabela

sorted_values = aggregated.sort_values(by='Passengers',ascending=False) # maior pro menor

sorted_values.head()
```

```python
# Cenario 1 : Agrupando por origem e destino 

grouped = df.groupby(['Origin', 'Destination'])

# Aplicar e agregar uma funcao ao grupo 

aggregated = grouped.agg({'Passengers':'sum','Flights':'sum'})

#Selecione os valores dessa tabela

sorted_values = aggregated.sort_values(by=['Passengers','Flights'],ascending=[False,False]) # maior pro menor

sorted_values.head()
```

```python
# Cenario 2 : Agrupamento por data de voo

#Convertendo Fly date em datetime e criando mes para agrupamento 

df['Month'] = pd.to_datetime(df['Fly Date'],format = '%Y%m').dt.to_period('M')

#Agrupar por mes e agregar a soma da coluna voos 

grouped = df.groupby('Month')
aggregated = grouped.agg({'Flights':'sum'})

# Classificando por mes 

sorted_values = aggregated.sort_values(by='Month')
sorted_values
```

# CUT

```python
# Criar grupos de intervalos para a distancia 

distance_bins = pd.cut(df['Distance'],bins= [0,500,1000,2000,3000,4000,df['Distance'].max()]) # bins significa que é um trecho.
```

```python
# Agrupe por intervalos de distancia e encontre o total de voos e passageiros para cada categoria.

grouped_data = df.groupby(distance_bins).agg({'Flights':'sum','Passengers':'sum'})

#Ordenando do maior pro menor 

sorted = grouped_data.sort_values(by='Distance',ascending = False)
sorted
```

```python
# Criando graficos 

# Agrupar por origem e agregando a soma dos voos. 
aeroportos = df.groupby('Origin').agg({'Flights':'sum'})

# Encontre os 10 aeroportos de origem mais movimentados
aeroportos_populares = aeroportos.nlargest(10,'Flights')

#Plote os dados 

aeroportos_populares.plot(kind = 'bar')
plt.title('Aeroportos mais populares')
plt.ylabel('Numero de voos')
plt.xticks(rotation = 45)
plt.tight_layout()
plt.show()
```

```python
# Cenario 2  # Identificando rotas + populares pela quantidade de passageios

trafego = df.groupby(['Origin','Destination']).agg({'Passengers':'sum'})
rotas_populares = trafego.nlargest(10,'Passengers')

rotas_populares.plot(kind='bar',color = 'purple')
plt.title('Rotas populares')
plt.ylabel('Numero de Passageiros')
plt.xticks(rotation = 45)
plt.tight_layout()
plt.show()
```
