```python
import pandas as pd
import numpy as np


data = {
	'Escola': ['A', 'B', 'A', 'B', 'A' , 'B' , 'A', 'B'],
	'Student ID': [1001, 1002, 1003, 1004, 1005, 1006,1007, 1008],
	'Matematica': np.random.randint(60,100,8),
	'Ingles': np.random.randint(60,100,8),
	'Historia': np.random.randint(60,100,8)
	}
```

```python
df = pd.DataFrame(data)
```

```python
print(df)
```

```python
agrupa = df.groupby('Escola')
```

```python
media_escola = agrupa.mean()
print(media_escola)
```

```python
#Aplicando multiplas operacoes

agg_data = agrupa.agg({
	'Matematica' : ['mean','median','std']}	)
print(agg_data)
```

* O desvio padrão nos diz que neste caso que quanto menor for, mais as notas são parecidas na escola B. Já na escola A o desvio está mais alto, o que significa que alunos podem ter tirado notas variadas entre altas e baixas
