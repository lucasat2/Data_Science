# Outliers - Valores muito fora da curva padrão.

IQR = Q3 - Q1 Uma regra comum:

* Valores menores que Q1 - 1.5*IQR → possíveis outliers inferiores.
* Valores maiores que Q3 + 1.5*IQR → possíveis outliers superiores.

Exemplo:

precos = df_loja["preco"]
q1 = precos.quantile(0.25)
q3 = precos.quantile(0.75)
iqr = q3 - q1
limite_inferior = q1 - 1.5 * iqr
limite_superior = q3 + 1.5 * iqr
limite_inferior, limite_superior

outliers = df_loja[(df_loja["preco"] < limite_inferior) | (df_loja["preco"] > limite_superior)]
outliers
