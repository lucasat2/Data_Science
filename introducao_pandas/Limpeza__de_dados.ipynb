```python
import numpy as np
import pandas as pd


# GERANDO VALORES NULOS
data = {
'A': [1,2,np.nan,4,5],
'B':[np.nan,2,3,4,5],
'C': [1,2,3,np.nan,np.nan]
}
```

```python
df = pd.DataFrame(data)
```

```python
df
```

```python
df.isna()  #tem valores nulos?
```

```python
df.dropna() # remove valores nulos das linhas e exibe só as linhas completas
```

```python
df.dropna(axis = 1) # remove colunas inteiras com pelo menos 1 valor nulo(prejudicial)
# Todas as colunas foram removidas pois tem pelo menos 1 valor nulo em todas.
```

```python
df.fillna("N") # substitui valores nulos por algo
```
