## Exercício 2 – Datas e tipos quebrados
Crie um DataFrame `df_pedidos` com as colunas:
* `id_pedido` (alguns como texto, ex.: "1", "2", "3")
* `data_pedido` (com formatos ligeiramente diferentes, ex.: "2024-10-01", "2024/10/02", "10-03-2024")
* `valor` (como texto, com vírgula e ponto misturados, e pelo menos um valor não numérico, ex.: "abc")

```python
import pandas as pd

dados_pedidos = {
    "id_pedido": ["1", "2", "3", "4", "5"],
    "data_pedido": ["2024-10-01", "2024/10/02", "10-03-2024", "2024-10-04", "data_errada"],
    "valor": ["200.50", "350,00", "150.00", "abc", "500.00"]
}
df_pedidos = pd.DataFrame(dados_pedidos)
df_pedidos
```

### Tarefa 2.1
Use `df_pedidos.dtypes` para ver os tipos atuais.

```python
df_pedidos.dtypes
```

### Tarefa 2.2
Converta `data_pedido` para datetime usando `pd.to_datetime` com `errors="coerce"`.
* Verifique se algum valor virou `NaT`.

```python
df_pedidos["data_pedido"] = pd.to_datetime(df_pedidos["data_pedido"], errors="coerce")
df_pedidos
```

### Tarefa 2.3
Crie uma coluna `valor_limpo` substituindo vírgula por ponto em `valor` (use `str.replace(",", ".", regex=False)`).

```python
df_pedidos['valor_limpo'] = df_pedidos['valor'].str.replace(",",".", regex=False)
df_pedidos
```

### Tarefa 2.4
Converta `valor_limpo` para numérico em uma nova coluna `valor_num` usando `pd.to_numeric` com `errors="coerce"`.

```python
df_pedidos['valor_num']=pd.to_numeric(df_pedidos['valor_limpo'],errors="coerce")
df_pedidos
```

### Tarefa 2.5
Remova as linhas em que `valor_num` ficou nulo, salvando o resultado em `df_pedidos_validos`.

```python
df_pedidos_validos = df_pedidos.dropna(subset=["valor_num"]).copy()
df_pedidos_validos
```

### Tarefa 2.6 (Opcional)
Converta `id_pedido` para inteiro com `astype("int64")`, se possível.

```python
df_pedidos_validos['id_pedido'] = df_pedidos_validos['id_pedido'].astype("int64")
df_pedidos_validos
```

Em texto, explique:
