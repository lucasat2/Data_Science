
# Exercício 1 – Vendas por loja

Crie, no seu notebook, um DataFrame de vendas com os dados abaixo (pode ser via dicionário ou lista de dicionários):

Colunas:

id_venda (1, 2, 3, 4)

loja ("Centro", "Shopping", "Centro", "Online")

produto ("Teclado", "Mouse", "Monitor", "Headset")

data ("2024-10-01", "2024-10-01", "2024-10-02", "2024-10-03")

valor (200.0, 150.0, 900.0, 300.0)

```python
import pandas as pd 

dados = {
    "id_venda": [1,2,3,4],
    "loja":["Centro","Shopping","Centro","Online"],
    "produto":["Teclado","Mouse","Monitor","Headset"],
    "data":["2024-10-01", "2024-10-01", "2024-10-02", "2024-10-03"],
    "valor":[200.0, 150.0, 900.0, 300.0]
}

df = pd.DataFrame(dados)
```

Tarefas:

Defina id_venda como índice usando set_index.

Converta data para datetime com pd.to_datetime.

Renomeie a coluna valor para valor_venda com rename.

Mostre o DataFrame final e verifique os tipos (dtypes).

```python
df_index_alterado = df.set_index("id_venda")
df_index_alterado
```

```python
df_index_alterado["datas"] = pd.to_datetime(df_index_alterado["data"])
df_index_alterado.dtypes
```

```python
df_renomeado = df_index_alterado.rename(columns={"valor": "valor_venda"})
df_renomeado
```

```python
df_renomeado.dtypes
```
