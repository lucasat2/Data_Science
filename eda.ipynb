## 2. O que é EDA (Exploratory Data Analysis)?
EDA = Exploratory Data Analysis = Análise Exploratória de Dados. Antes de treinar modelo, fazer dashboard ou tomar decisão, você precisa:

* Ver como os dados estão distribuídos.
* Entender a ordem de grandeza dos valores.
* Descobrir se tem dados faltando, categorias estranhas, outliers bizarros.
* Ter uma “sensação” do dataset: o que é normal, o que é raro, o que parece errado.

Pandas é perfeito para esse trabalho de detetive inicial.

Vamos trabalhar com um exemplo de uma loja de roupas.

````python
import pandas as pd

dados = {
    "produto": [
        "Camiseta Básica", "Calça Jeans", "Vestido Floral", "Jaqueta Couro",
        "Camiseta Básica", "Calça Jeans", "Shorts Jeans", "Moletom", "Moletom", "Vestido Floral"
    ],
    "categoria": [
        "Camisetas", "Calças", "Vestidos", "Jaquetas",
        "Camisetas", "Calças", "Shorts", "Moletons", "Moletons", "Vestidos"
    ],
    "preco": [49.90, 159.90, 199.90, 399.90, 59.90, 149.90, 89.90, 129.90, 139.90, 219.90],
    "quantidade_vendida": [120, 80, 45, 10, 150, 70, 60, 90, 110, 30]
}

df_loja = pd.DataFrame(dados)
df_loja
````

Dá uma olhada rápida:

````python
df_loja.info()
````

## 6. Quartis, mínimos e máximos (`min`, `max`, `quantile`)
Além do `describe()`, às vezes queremos olhar especificamente para certos pontos.

### 6.1. Mínimo, máximo e média

````python
df_loja["preco"].min()
````

````python
df_loja["preco"].max()
````

````python
df_loja["preco"].mean()
````

### 6.2. Quartis com quantile

````python
q1 = df_loja["preco"].quantile(0.25)
q2 = df_loja["preco"].quantile(0.50)  # mediana
q3 = df_loja["preco"].quantile(0.75)

q1, q2, q3
````

Isso ajuda a entender:

* Até quanto estão os 25% mais baratos (q1).
* “Preço típico” (mediana, q2).
* Faixa dos 25% mais caros (q3 até o máximo).

## 7. Identificação inicial de outliers (intuitivamente)
Outlier = valor muito fora da curva. Exemplos:

* Uma jaqueta de R$ 10.000,00 no meio de produtos de R$ 50 a R$ 400.
* Uma quantidade vendida de 10.000 unidades onde o normal é 50–200.

### 7.1. Inspeção visual com describe

````python
df_loja["preco"].describe()
````

Se o max for muuuuito maior que o 75%, é sinal de que tem algo estranho (ou um produto premium, ou erro).

<a id="secao-outliers"></a>

### 7.2. IQR (Intervalo Interquartílico) – versão simples
IQR = Q3 - Q1 Uma regra comum:

* Valores menores que Q1 - 1.5*IQR → possíveis outliers inferiores.
* Valores maiores que Q3 + 1.5*IQR → possíveis outliers superiores.

Exemplo:

````python
precos = df_loja["preco"]

q1 = precos.quantile(0.25)
q3 = precos.quantile(0.75)
iqr = q3 - q1

limite_inferior = q1 - 1.5 * iqr
limite_superior = q3 + 1.5 * iqr

limite_inferior, limite_superior
````

Depois, você pode ver quais produtos fogem muito:

````python
outliers = df_loja[
    (df_loja["preco"] < limite_inferior) | (df_loja["preco"] > limite_superior)]
outliers
````

Na nossa tabelinha pequena é provável que não haja outliers bizarros, mas é assim que você faria num dataset real. ⚠️ Importante: Outlier não é “erro” automaticamente. Pode ser:

* Um produto premium.
* Uma promoção muito louca.
* Um erro de digitação (R$ 10.000 em vez de R$ 100).

EDA serve pra levantar suspeitas, não pra tirar conclusões definitivas sozinho.

## 9. Segundo dataset exemplo: Plataforma de Cursos Online
Vamos agora para um contexto de plataforma de cursos.

````python
dados_cursos = {
    "curso": [
        "Python Básico", "Python Avançado", "Data Science com Python",
        "Introdução ao SQL", "Machine Learning", "Excel para Dados",
        "Power BI Essencial", "Estatística Aplicada"
    ],
    "area": [
        "Programação", "Programação", "Data Science",
        "Banco de Dados", "Data Science", "Ferramentas",
        "Ferramentas", "Estatística"
    ],
    "avaliacao": [4.5, 4.7, 4.8, 4.2, 4.6, 4.3, 4.4, 4.1],
    "horas": [12, 20, 25, 10, 30, 8, 15, 18],
    "qtde_alunos": [1200, 800, 950, 600, 700, 1100, 900, 500]
}

df_cursos = pd.DataFrame(dados_cursos)
df_cursos
````

### 9.1. Distribuição por área

````python
df_cursos["area"].value_counts()
````

Isso te mostra quais áreas têm mais cursos.

### 9.2. Estatísticas das colunas numéricas

````python
df_cursos[["avaliacao", "horas", "qtde_alunos"]].describe()
````

Perguntas que você pode responder:

* Qual a avaliação média dos cursos?
* Qual curso é mais longo?
* Quantos alunos em média cada curso tem?

### 9.3. Quem tem mais alunos? Quem é melhor avaliado?

````python
df_cursos.sort_values("qtde_alunos", ascending=False).head(1)
````

````python
df_cursos.sort_values("avaliacao", ascending=False).head(1)
````

Com isso você consegue escrever:

> “O curso mais popular em número de alunos é X, com Y alunos. O curso com melhor avaliação é Z, com nota W.”

Ou ainda:

> “A área de Data Science tem cursos bem avaliados e com boa carga horária, sendo uma área relevante para investir.”

Esse tipo de frase é exatamente o que você leva pra um relatório ou apresentação.
