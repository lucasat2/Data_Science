```python
import requests
import pandas as pd
from pandas import json_normalize



url = "https://jsonplaceholder.typicode.com/users"

#Puxando os dados da url
resposta = requests.get(url)

resposta.raise_for_status()

#Converter o resultado da url em json
dados = resposta.json()  

# Achata o JSON
df = json_normalize(dados)


print("Colunas disponíveis:")
print(df.columns)


# Vamos escolher algumas colunas
colunas_interesse = [
    "id",
    "name",
    "email",
    "address.city",
    "address.street",
    "company.name"
]

df_simplificado = df[colunas_interesse].rename(
    columns={
        "name": "nome",
        "address.city": "cidade",
        "address.street": "rua",
        "company.name": "empresa"
    }
)

print("\nDataFrame simplificado:")
print(df_simplificado.head())
print("\nInfo:")
print(df_simplificado.info())
```
