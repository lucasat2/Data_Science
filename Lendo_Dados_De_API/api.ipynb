```python
import requests

url = "https://jsonplaceholder.typicode.com/posts/1"  # API pública de teste
resposta = requests.get(url)

print("Status code:", resposta.status_code)
print("Corpo da resposta (texto):")
print(resposta.text)
```

 Convertendo a resposta para JSON


```python
url = "https://jsonplaceholder.typicode.com/posts/1"
resposta = requests.get(url)

print("Status code:", resposta.status_code)

# Converter o corpo em JSON (dicionário Python)
dados = resposta.json()

print("Tipo de 'dados':", type(dados))
print("Conteúdo de 'dados':", dados)
print("Título do post:", dados["title"])
```

Entendendo melhor o objeto Response

```python
resposta = requests.get(url)
print(resposta.status_code)   # código HTTP (200, 404, 500...)
print(resposta.headers)       # cabeçalhos da resposta (tipo de conteúdo, servidor, etc.)
print(resposta.url)           # URL final (às vezes diferente se houve redirecionamento)
print(resposta.text)          # corpo como texto
# resposta.json()             # corpo como JSON (se aplicável)
```

### **Por que isso importa para Data Science?**

- `status_code` → garante que você só continua o processamento quando a requisição deu certo.
- `headers` → às vezes carregam informações como tipo de conteúdo (`Content-Type`), paginação, limites de uso de API etc.
- `url` → ajuda no debug, especialmente quando você monta URLs com parâmetros.

## **7. Enviando parâmetros na URL com `params`**

Muitas APIs usam **parâmetros de consulta** (query string) para filtrar ou personalizar os dados retornados. Exemplo de URL com parâmetros:

https://api.exemplo.com/produtos?categoria=livros&ordenar=preco

Você não precisa montar essa URL manualmente. O requests faz isso pra você usando o argumento params.

```python
url = "https://jsonplaceholder.typicode.com/comments"

parametros = {
    "postId": 1  # queremos comentários apenas do post com id=1
}

resposta = requests.get(url, params=parametros)

print("URL final:", resposta.url)  # veja como ficou a query string
print("Status code:", resposta.status_code)

comentarios = resposta.json()
print("Quantidade de comentários retornados:", len(comentarios))
print("Primeiro comentário:", comentarios[0])
print("Segundo comentário:", comentarios[1])
```

# Tratando erros e situações comuns

```python
import requests

url = "https://jsonplaceholder.typicode.com/posts/1"
resposta = requests.get(url)

if resposta.status_code == 200:
    dados = resposta.json()
    print("Título:", dados["title"])
else:
    print("Falha ao buscar dados. Código:", resposta.status_code)
```

Você pode generalizar:

- `2xx` → sucesso.
- `4xx` → erro do cliente (URL errada, parâmetro inválido, não autorizado).
- `5xx` → erro no servidor (problema do lado da API).
