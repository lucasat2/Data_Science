## Exercício 1 – Faturamento por cidade
Use o DataFrame `df_vendas` (ou crie um semelhante) com as colunas:
* `id_venda` 
* `cidade` 
* `canal_venda` 
* `valor` 

Com pelo menos 10–15 linhas, misturando cidades e canais.

```python
import pandas as pd

dados_vendas = {
    "id_venda": range(1, 16),
    "cidade": [
        "São Paulo", "Rio de Janeiro", "Belo Horizonte", "São Paulo", "Curitiba",
        "Rio de Janeiro", "Belo Horizonte", "São Paulo", "Curitiba", "Porto Alegre",
        "São Paulo", "Rio de Janeiro", "Belo Horizonte", "Curitiba", "Porto Alegre"
    ],
    "canal_venda": [
        "Online", "Loja Física", "Online", "Marketplace", "Online",
        "Online", "Loja Física", "Online", "Marketplace", "Loja Física",
        "Marketplace", "Loja Física", "Online", "Online", "Marketplace"
    ],
    "valor": [250.0, 450.0, 300.0, 150.0, 600.0, 200.0, 800.0, 500.0, 350.0, 400.0, 300.0, 900.0, 150.0, 450.0, 250.0]
}

df_vendas = pd.DataFrame(dados_vendas)
df_vendas
```

### Tarefa 1.1
Calcule o valor total por cidade usando `groupby`.
* Salve o resultado em `faturamento_por_cidade`.
* Transforme em DataFrame com `reset_index()`.

```python
faturamento_por_cidade = df_vendas.groupby('cidade').agg({'valor':'sum'}).reset_index()
faturamento_por_cidade
```
