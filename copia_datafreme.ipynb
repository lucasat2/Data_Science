# Cópia vs “view” e o SettingWithCopyWarning

# O problema clássico

Quando você filtra um DataFrame, às vezes o pandas não tem certeza se você está mexendo na visão (view) do original ou em uma cópia independente.

```python
import pandas as pd 

datas = pd.date_range(start="2024-10-01", periods=10, freq="D")

dados_vendas = {
    "data": datas,
    "loja": ["Centro", "Centro", "Shopping", "Shopping", "Centro",
             "Centro", "Shopping", "Centro", "Shopping", "Centro"],
    "faturamento": [150, 200, 400, 500, 800, 200, 200, 100, 480, 301]
}
df_vendas = pd.DataFrame(dados_vendas)
df_vendas
```

Existem duas boas práticas:
# Usar .loc no DataFrame original

```python
df_vendas.loc[df_vendas["faturamento"] > 300, "valor_com_desconto"] = df_vendas["faturamento"] * 0.9
df_vendas
```

```python
df_vendas.fillna(value="Sem desconto")
```

# Criar uma cópia explícita
Se você realmente quer trabalhar em um DataFrame separado:

Ao usar .copy(), você diz explicitamente:

“Sim, eu quero um DataFrame independente.”

```python
df_caros = df_vendas[df_vendas["faturamento"] > 300].copy()

df_caros["valor_com_desconto"] = df_caros["faturamento"] * 0.9

df_caros
```
