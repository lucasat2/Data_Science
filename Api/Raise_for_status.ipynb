# Usando raise_for_status()
O método raise_for_status() levanta uma exceção se o código não for 2xx:

```python
import requests

url = "https://jsonplaceholder.typicode.com/posts/1"
resposta = requests.get(url)

try:
    resposta.raise_for_status()  # se não for 2xx, lança erro
    dados = resposta.json()
    print("Título:", dados["title"])
except requests.exceptions.HTTPError as e:
    print("Ocorreu um erro HTTP:", e)
```

Isso é útil em scripts mais robustos, onde você quer “falhar cedo” se algo estiver errado.
