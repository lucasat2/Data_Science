```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
```

```python
df = pd.read_csv('flights.csv')
```

# 1

```python
cidade_alvo = 'Eugene, OR'
total_voos = df.loc[df['Origin City'] == cidade_alvo, 'Flights'].sum()
print(f"Total de voos saindo de {cidade_alvo}: {total_voos}")
```

# 2

```python
media_origem = df.groupby('Origin').agg({'Distance':'mean'})
media_origem = df.groupby('Origin State').agg({'Distance':'mean'})
media_origem 
```

# 3

```python
mais_receberam = df.groupby('Destination City').agg({'Passengers':'sum'}).nlargest(5, 'Passengers')
mais_receberam.plot(kind='bar')
plt.title('Cidades que mais receberam passageiros')
plt.xlabel('Cidade de destino')
plt.ylabel('Passageiros')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
