```python
import pandas as pd

dados_vendas = {
    "id_venda": range(1, 16),
    "cidade": [
        "São Paulo", "Rio de Janeiro", "Belo Horizonte", "São Paulo", "Curitiba",
        "Rio de Janeiro", "Belo Horizonte", "São Paulo", "Curitiba", "Porto Alegre",
        "São Paulo", "Rio de Janeiro", "Belo Horizonte", "Curitiba", "Porto Alegre"
    ],
    "canal_venda": [
        "Online", "Loja Física", "Online", "Marketplace", "Online",
        "Online", "Loja Física", "Online", "Marketplace", "Loja Física",
        "Marketplace", "Loja Física", "Online", "Online", "Marketplace"
    ],
    "valor": [250.0, 450.0, 300.0, 150.0, 600.0, 200.0, 800.0, 500.0, 350.0, 400.0, 300.0, 900.0, 150.0, 450.0, 250.0]
}

df_vendas = pd.DataFrame(dados_vendas)
df_vendas
```

### Renomeando colunas com rename

```python
df_renomeado = df_vendas.rename(columns={"valor": "valor_venda"})
print(df_renomeado)
```

Você também pode renomear mais de uma vez:



```python
df_renomeado = df_vendas.rename(columns={
    "loja": "nome_loja",
    "produto": "nome_produto"
})
print(df_renomeado)
```

Boa prática:
Use nomes de colunas consistentes, tudo minúsculo, sem acento e com _ em vez de espaço.
Ex.: valor_total, data_pedido, cidade_cliente.
