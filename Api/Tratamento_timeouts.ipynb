# Tratando timeouts (API demorando a responder)

Às vezes a API demora demais ou está instável. Você não quer que seu script fique “preso” indefinidamente. Use o parâmetro timeout:

```python
import requests

url = "https://jsonplaceholder.typicode.com/posts/1"

try:
    resposta = requests.get(url, timeout=5)  # espera no máximo 5 segundos
    resposta.raise_for_status()
    dados = resposta.json()

    print("Título:", dados["title"])
    
except requests.exceptions.Timeout:
    print("A requisição demorou demais e foi cancelada (timeout).")
except requests.exceptions.RequestException as e:
    print("Erro ao fazer requisição:", e)
```

Aqui usamos uma exceção mais genérica RequestException para capturar possíveis problemas de conexão, DNS, etc.
