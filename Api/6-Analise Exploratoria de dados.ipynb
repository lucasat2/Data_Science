```python
import pandas as pd
import numpy as np

dados = [
    {"id": 1, "cliente": " Ana ", "valor": "250.0", "data": "2025-11-01", "estado": "mg"},
    {"id": 2, "cliente": "Bruno", "valor": "300",   "data": "2025/11/02", "estado": "SP"},
    {"id": 3, "cliente": "Carla", "valor": None,    "data": "2025-11-03", "estado": "rj"},
    {"id": 4, "cliente": "Ana",   "valor": "abc",   "data": None,         "estado": "MG"},
    {"id": 4, "cliente": "Ana",   "valor": "abc",   "data": None,         "estado": "MG"}
]

df = pd.DataFrame(dados)

# 1) Remover duplicatas
df = df.drop_duplicates()

# 2) Limpar textos
df["cliente"] = df["cliente"].str.strip()
df["estado"] = df["estado"].str.upper()

# 3) Converter tipos
df["valor"] = pd.to_numeric(df["valor"], errors="coerce")
df["data"] = pd.to_datetime(df["data"], errors="coerce")

# 4) Tratar valores ausentes (exemplo simples)
#    - se valor for NaN, substitui por 0
#    - se data for NaT, deixa como está (poderíamos decidir outra coisa)
df["valor"] = df["valor"].fillna(0)

# 5) Criar coluna de desconto (10%)
df["valor_com_desconto"] = df["valor"] * 0.9

# 6) Criar faixa de valor
condicoes = [
    df["valor"] < 200,
    (df["valor"] >= 200) & (df["valor"] < 500),
    df["valor"] >= 500
]
faixas = ["Baixo", "Médio", "Alto"]

df["faixa_valor"] = np.select(condicoes, faixas, default="Desconhecido")

print(df)
print("\nInfo:")
print(df.info())
```
