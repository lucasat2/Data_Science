Exercicio 1 - Você recebeu a seguinte estrutura (simulando resposta de uma API):

```python
usuarios = [
    {"id": 1, "nome": "Ana",   "idade": 28, "cidade": "Juiz de Fora"},
    {"id": 2, "nome": "Bruno", "idade": 22, "cidade": "São Paulo"},
    {"id": 3, "nome": "Carla", "idade": 31, "cidade": "Rio de Janeiro"},
    {"id": 4, "nome": "Diego", "idade": 19, "cidade": "Juiz de Fora"}
]
```

Tarefas:

Percorra a lista e imprima, para cada usuário, uma frase no formato:
"Ana (28 anos) - Juiz de Fora"

Crie uma nova lista chamada usuarios_jf contendo apenas os usuários cuja cidade é "Juiz de Fora".

Imprima quantas pessoas moram em Juiz de Fora e os nomes delas.

Dica: use for e if.

```python
usuarios_jf = []

print("Total de usuarios da lista")
for u in usuarios:
    print(f'{u["nome"]} ({u["idade"]} anos) - {u["cidade"]}')
    
    if u["cidade"] == "Juiz de Fora":
        usuarios_jf.append(u)


print("\nSomente os moradores de Juiz de Fora")
for jf in usuarios_jf:
    print(jf["nome"])

print("\n Quantidade de pessoas em Juiz de Fora:", len(usuarios_jf))

```

# Exercício 2 – Loja online com JSON aninhado

```python
loja = {
    "nome": "Codi Tech Store",
    "produtos": [
        {
            "id": 1,
            "nome": "Notebook",
            "preco": 3500.0,
            "categorias": ["eletrônicos", "computadores"],
            "disponivel": True
        },
        {
            "id": 2,
            "nome": "Headset Gamer",
            "preco": 300.0,
            "categorias": ["eletrônicos", "áudio"],
            "disponivel": False
        },
        {
            "id": 3,
            "nome": "Webcam Full HD",
            "preco": 200.0,
            "categorias": ["eletrônicos", "vídeo"],
            "disponivel": True
        }

        ,
        {
            "id": 4,
            "nome": "Impressora",
            "preco": 250.0,
            "categorias": ["eletrônicos", "escritório"],
            "disponivel": False
        }

        
    ]
}
```

Tarefas:

Imprima o nome da loja.

Percorra a lista de produtos e imprima uma linha com:
"Notebook - R$3500.00 - disponível" ou "Headset Gamer - R$300.00 - indisponível"
(use "disponível" se disponivel == True, caso contrário "indisponível").

Crie uma lista chamada produtos_disponiveis contendo apenas o nome dos produtos com disponivel == True.

(Desafio extra) Calcule o preço médio dos produtos disponíveis.

```python
print(loja["nome"])
```

```python
produtos_disponiveis= []
precos_disponiveis=[]

for produto in loja["produtos"]:
    nome=produto["nome"]
    preco=produto["preco"]
    
    if produto["disponivel"] == True:
        situacao = "Disponivel"
        produtos_disponiveis.append(nome)
        precos_disponiveis.append(preco)
        
    else:
        situacao = "Indisponivel"
    print("Nome Produto:",nome,"-","Preço:",preco,"-","Situacao:",situacao)
    

#PRODUTOS DISPONIVEIS
print("\nProdutos Disponiveis",produtos_disponiveis)

#MEDIA DOS PRODUTOS DISPONIVEIS
media = sum(precos_disponiveis)/ len(precos_disponiveis)
print(f"Preço medio dos disponiveis: RS{media:.2f}")
```
