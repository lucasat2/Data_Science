# 1. Objetivos da aula
Ao final desta aula, você será capaz de:

* Selecionar colunas específicas de um DataFrame.
* Selecionar linhas por posição e por rótulo usando `iloc` e `loc`.
* Criar filtros booleanos para selecionar linhas que atendem a certas condições.
* Combinar condições com:
    * `&` → E
    * `|` → OU
    * `~` → NÃO
* Evitar erros comuns com parênteses em filtros.

Essa é a base da análise: escolher só os dados certos pra responder à pergunta certa.

## 2. Dataset exemplo: Clientes de um banco
Vamos montar um DataFrame de exemplo que usaremos ao longo da aula:

```
import pandas as pd

dados_clientes = {
    "id_cliente": [1, 2, 3, 4, 5],
    "nome": ["Ana", "Bruno", "Carla", "Daniel", "Eduarda"],
    "idade": [22, 35, 29, 42, 31],
    "cidade": ["Juiz de Fora", "São Paulo", "Juiz de Fora", "Rio de Janeiro", "São Paulo"],
    "renda_mensal": [2500.0, 7000.0, 4800.0, 12000.0, 5200.0]
}

df_clientes = pd.DataFrame(dados_clientes)
df_clientes
```

## 3. Selecionando colunas

### 3.1. Uma única coluna

```
df_clientes["nome"]
```

Retorna uma Series (uma coluna).
Tipo: `pandas.Series`.
Se quiser ver os tipos:

```
type(df_clientes["nome"])
```

### 3.2. Várias colunas
Para pegar mais de uma, use uma lista de nomes:

```
df_clientes[["nome", "cidade"]]
```

Retorna um DataFrame com apenas essas colunas.
Você pode usá-lo para mostrar só as informações importantes numa análise.

## 4. Selecionando linhas por posição: `iloc` 
`iloc` trabalha com posição numérica (estilo “linha 0, linha 1, linha 2…”).

### 4.1. Uma linha específica

```
df_clientes.iloc[0]   # primeira linha
```

```
df_clientes.iloc[3]   # quarta linha
```

### 4.2. Várias linhas por fatiamento (slice)

```
df_clientes.iloc[0:3]   # linhas 0, 1 e 2
```

O intervalo é `[início : fim)`, ou seja, inclui o início e exclui o fim.

### 4.3. Linhas e colunas ao mesmo tempo
`df.iloc[linhas, colunas]` Exemplos:

```
# Linhas 0 a 2, colunas 0 a 2 (id_cliente, nome, idade)
df_clientes.iloc[0:3, 0:3]
```

```
# Apenas a coluna "idade" (índice 2) das 3 primeiras linhas
df_clientes.iloc[0:3, 2]
```

## 5. Selecionando linhas por rótulo: `loc` 
`loc` trabalha com rótulos (nome das colunas e índice das linhas). Como nosso índice ainda é o padrão (0,1,2,…), fica muito parecido com `iloc` nas linhas. Mas a vantagem do `loc` é que podemos:

* Passar condições booleanas.
* Passar nomes de colunas diretamente.

### 5.1. Linhas por índice + colunas por nome

```
# Todas as colunas da linha com índice 0
df_clientes.loc[0]
```

```
# Linhas 0 a 2, só as colunas "nome" e "cidade"
df_clientes.loc[0:2, ["nome", "cidade"]]
```

Repare que com `loc`, o fim do slice de linha é incluído:

`loc[0:2]` → inclui 0, 1, 2.

## 6. Filtros booleanos: selecionando linhas por condição
Um filtro booleano é uma expressão que devolve `True`/`False` para cada linha. Exemplo: “clientes com renda acima de 5000”.

```
filtro_renda = df_clientes["renda_mensal"] > 5000
filtro_renda
```

Você passa esse filtro dentro de `[]`:

```
df_clientes[filtro_renda]
```

Ou direto:

```
df_clientes[df_clientes["renda_mensal"] > 5000]
```

### 6.1. Guardando o filtro em variável (boa prática)
É comum fazer:

```
filtro_ricos = df_clientes["renda_mensal"] > 8000
clientes_ricos = df_clientes[filtro_ricos]
clientes_ricos
```

Assim o código fica mais legível, e você pode reutilizar o filtro em outros contextos.

## 7. Combinando condições: `&`, `|` e `~` 
Agora vem a cereja do bolo: filtros mais complexos.

* `&` → E (and)
* `|` → OU (or)
* `~` → NÃO (not)

⚠️ **IMPORTANTE:** Você precisa de parênteses em cada condição, senão o Python se confunde com a ordem das operações.

### 7.1. E lógico (&)
“Clientes com renda > 5000 E idade > 30”:

```
filtro = (df_clientes["renda_mensal"] > 5000) & (df_clientes["idade"] > 30)
df_clientes[filtro]
```

### 7.2. OU lógico (|)
“Clientes de Juiz de Fora OU São Paulo”:

```
filtro = (df_clientes["cidade"] == "Juiz de Fora") | (df_clientes["cidade"] == "São Paulo")
df_clientes[filtro]
```

### 7.3. NÃO lógico (~)
“Clientes que não são de São Paulo”:

```
filtro_nao_sp = ~(df_clientes["cidade"] == "São Paulo")
df_clientes[filtro_nao_sp]
```

Uma forma equivalente:

```
df_clientes[df_clientes["cidade"] != "São Paulo"]
```

## 8. Combinando `loc` com filtros e seleção de colunas
Você pode juntar tudo:

1. Filtro de linhas.
2. Seleção de colunas.

Exemplo: “Quero só `id_cliente` e `cidade` para clientes com `idade > 30` e `renda > 5000”.

```
filtro = (df_clientes["idade"] > 30) & (df_clientes["renda_mensal"] > 5000)

df_clientes.loc[filtro, ["id_cliente", "cidade"]]
```

Isso é muito comum na prática:

`loc[condição, ["col1", "col2", ...]]`

## 9. Outro dataset exemplo: Estoque de Supermercado
Vamos criar um DataFrame de estoque pra praticar filtros:

```
dados_estoque = {
    "produto": [
        "Maçã", "Banana", "Leite Integral", "Arroz 5kg", "Feijão 1kg",
        "Tomate", "Detergente", "Sabão em Pó", "Refrigerante", "Alface"
    ],
    "setor": [
        "Hortifruti", "Hortifruti", "Frios", "Mercearia", "Mercearia",
        "Hortifruti", "Limpeza", "Limpeza", "Bebidas", "Hortifruti"
    ],
    "preco": [6.99, 4.50, 4.79, 25.90, 7.99, 8.50, 3.49, 18.90, 7.50, 4.20],
    "estoque": [50, 120, 30, 80, 60, 15, 200, 40, 90, 10]
}

df_estoque = pd.DataFrame(dados_estoque)
df_estoque
```

### 9.1. Filtro simples: setor Hortifruti

```
filtro_horti = df_estoque["setor"] == "Hortifruti"
df_estoque[filtro_horti]
```

### 9.2. Filtro composto: Hortifruti com estoque baixo
“Hortifruti com estoque < 20”:

```
filtro = (df_estoque["setor"] == "Hortifruti") & (df_estoque["estoque"] < 20)
df_estoque[filtro]
```

### 9.3. Produtos baratos OU com muito estoque

```
filtro = (df_estoque["preco"] < 5) | (df_estoque["estoque"] > 100)
df_estoque[filtro]
```

### 9.4. Selecionando só algumas colunas nesse filtro

```
df_estoque.loc[filtro, ["produto", "preco", "estoque"]]
```
