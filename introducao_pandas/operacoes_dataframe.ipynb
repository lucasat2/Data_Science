```python
import numpy as np
import pandas as pd

# CRIAR UMA SERIE COM NUMEROS ALEATORIOS

data_series = np.random.normal(0, 1 ,100)
data_series2 = np.random.normal(0,1,100)

frame = {
'coluna1': data_series,
'coluna2': data_series2

}
df = pd.DataFrame(frame)

df
```

```python
df.describe()
```

```python
df.head() # 5 primeiros registros
```

```python
df.tail()  # 5 ultimos registros
```

```python
df.sort_values(by='coluna1') # classificar valores
```

```python
df['Total'] = df['coluna1'] + df['coluna2'] # criar nova coluna com soma de outras 2
```

```python
df
```

```python
df.drop(columns='Total', inplace= True) # Apagar uma coluna
```

```python
data = {
	'Escola': ['A', 'B', 'A', 'B', 'A' , 'B' , 'A', 'B'],
	'Student ID': [1001, 1002, 1003, 1004, 1005, 1006,
	1007, 1008],
	'Matematica': np.random.randint(60,100,8),
	'Ingles': np.random.randint(60,100,8),
	'Historia': np.random.randint(60,100,8)
	}
```

```python
df2 = pd.DataFrame(data)
```

```python
df2
```

```python
df2[df2['Matematica']>75] # condicional
```

```python
df2.loc[5:10,['Matematica','Ingles']] # fatiamento de coluna
```

```python
df2.loc[(df2['Matematica']>75) | (df2['Ingles'] < 60), ['Matematica','Ingles']] # fatiamento com operador logico
```

```python
df2['total'] = df2['Matematica'] + df2['Ingles'] + df2['Historia']
```
